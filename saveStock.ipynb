# 存股族必看！ 殖利率計算
殖利率公式:   $$殖利率 = \frac{股息}{除權息前股價}$$

## Part 1: 下載證交所除權除息結果表
資料來源:[證交所 除權除息計算結果表](https://www.twse.com.tw/zh/page/trading/exchange/TWT49U.html)

https://colab.research.google.com/drive/12-UKGXNrwWg-2e8It5r3yxyZCBvocIWr?usp=sharing#scrollTo=TxRpnZy8iKqs

```python
import requests as r
url = 'https://www.twse.com.tw/rwd/zh/exRight/TWT49U?startDate=20230101&endDate=20231231&response=json&_=1703753307644'
r.get(url)
res = r.get(url)
stock_json = res.json()
stock_json['data']
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(stock_json['data'])
df
```

```python
stock_json['fields']
```

```python
df.columns = stock_json['fields']
df
```

## Part 2: 資料整理
#### 2.1 選取所需欄位

```python
df = df[['資料日期', '股票代號', '股票名稱', '除權息前收盤價', '除權息參考價', '權值+息值', '權/息']]
df
```

#### 2.2 重新命名欄位

```python
df.columns = ['資料日期', '股票代號', '股票名稱', '除權息前收盤價', '除權息參考價', '權值_息值', '權_息']
df
```

#### 2.3 資料轉型

```python
for row in range(len(df)):
    for col in range(3, 6):
        if type(df.iloc[row, col]) == str:
            df.iloc[row, col] = float(df.iloc[row, col].replace(',', ''))
        else:
            df.iloc[row, col] = float(df.iloc[row, col])
df
```

## Part 3: 殖利率計算

```python
df['殖利率%'] = (df.權值_息值 / df.除權息前收盤價) * 100
df
```

```python
df.sort_values(by = ['殖利率%'], ascending = False)
```
